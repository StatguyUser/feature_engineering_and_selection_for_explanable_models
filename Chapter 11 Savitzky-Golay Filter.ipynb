```python
import pandas as pd
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:\\Book\\beaver1.csv')
```

```python
data['savitzky-golay'] = savgol_filter(data['temp'],polyorder=5,window_length=11)
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.style.use('ggplot') 

plt.plot(data.index,data['temp'],linestyle = None,marker='.',label = 'Original Temperature')
plt.plot(data.index,data['savitzky-golay'],linestyle = None,marker='.',label = 'Filtered Temperature')
plt.xlabel('Observations', fontsize=13)
plt.ylabel('Temperature', fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(fontsize=15)
```
