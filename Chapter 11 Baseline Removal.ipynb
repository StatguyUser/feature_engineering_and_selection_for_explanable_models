```python
import pandas as pd
from BaselineRemoval import BaselineRemoval
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:\\Book\\spectra_49999_94078628963_milksample.csv')
```

```python
baseObj = BaselineRemoval(data['Spectra1'])
```

```python
data['Modpoly'] = baseObj.ModPoly()
data['Imodpoly'] = baseObj.IModPoly()
data['Zhang (airPLS)'] = baseObj.ZhangFit()
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.style.use('ggplot') 

plt.plot(data.index,data['Spectra1'],linestyle = None,marker='.',label = 'Original')
plt.plot(data.index,data['Modpoly'],linestyle = None,marker='.',label = 'Modpoly')
plt.xlabel('Observations', fontsize=13)
plt.ylabel('Spectra', fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(fontsize=15)
```
